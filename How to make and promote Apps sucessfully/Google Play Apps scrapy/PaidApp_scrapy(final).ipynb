```python
from selenium import webdriver
browser = webdriver.Chrome()
import time
```

```python
session = browser.find_elements_by_css_selector('.card-content.id-track-click.id-track-impression')
url_list = []
for card in session:
    url = card.find_element_by_css_selector('.card-click-target')
    url = url.get_attribute('href')
    url_list.append(url)
```

```python
len(url_list)
```

```python
all_info=[]
for url in url_list:
    r=requests.get(url).text
    data=BeautifulSoup(r,'html.parser')
    info={}
    details1=data.find_all('span',attrs={'class':'htlgb'})
    details2=data.find_all('a',attrs={'class':'hrTbp R8zArc'})
    info['title'] =data.find('h1').text.strip()
    #info['score']=data.find('div',attrs={'class':'BHMmbe'}).text.strip()
    info['date']=details1[2].text
    info['size']=details1[4].text
    info['installs']=details1[6].text
    info['content_rating']=details1[12].div.div.text
    info['category']=details2[1].text
    try:
        info['score']=data.find('div',attrs={'class':'BHMmbe'}).text.strip()
    except:
        pass
    all_info.append(info)
all_info
```

```python
import pandas as pd
import numpy as np
```

```python
df_paid=pd.DataFrame(all_info)
```
