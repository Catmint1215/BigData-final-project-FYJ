```python
from selenium import webdriver
browser = webdriver.Chrome()
import time
```

```python
browser.get('https://play.google.com/store/apps/collection/topselling_free')
for i in range(9):
    time.sleep(2)
    if i <3:
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    elif i == 3:
        browser.find_element_by_css_selector('#show-more-button').click()
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    else:
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')

#     article_urls=[]
    
#     for my_url in my_urls:
#         url='{0}{1}'.format('https://play.google.com',my_url['href'])
#         article_urls.append(url)
#     return article_urls
# url_list=scrape_articles_urls_of_one_page('https://play.google.com/store/apps/collection/topselling_free')
```

```python
session = browser.find_elements_by_css_selector('.card-content.id-track-click.id-track-impression')
url_list = []
for card in session:
    url = card.find_element_by_css_selector('.card-click-target')
    url = url.get_attribute('href')
    url_list.append(url)
```

```python
url_list 
```

```python
len(url_list)
```

```python
all_info=[]
for url in url_list:
    r=requests.get(url).text
    data=BeautifulSoup(r,'html.parser')
    info={}
    details1=data.find_all('span',attrs={'class':'htlgb'})
    details2=data.find_all('a',attrs={'class':'hrTbp R8zArc'})
    info['title'] =data.find('h1').text.strip()
    info['score']=data.find('div',attrs={'class':'BHMmbe'}).text.strip()
    info['date']=details1[0].text
    info['size']=details1[2].text
    info['installs']=details1[4].text
    info['content_rating']=details1[10].div.div.text
    info['category']=details2[1].text
    all_info.append(info)
all_info
```

```python
import pandas as pd
import numpy as np
```

```python
df_free=pd.DataFrame(all_info)
```
